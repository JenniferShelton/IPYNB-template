# Lokta-Volterra Ecology Modeling

The Lotka-Volterra Model demonstrates the relationship between two populations of organisms, one the predator and the other the prey. The entire system is modeled through two first-order ordinary differential equations (ODEs). These represent the overall growth rates of the two populations by incorporating various constants (we assume prey have unlimited food and reproduce at a constant rate, while the predator appetite is effectively unlimited given sufficient prey).

This example will implement numpy, pylab, and scipy packages and is based off this wonderful tutorial: (http://wiki.scipy.org/Cookbook/LoktaVolterraTutorial).

```python
# du/dt = a*u - b*u*v      this is the eqn for the prey population
# dv/dt = -c*v + d*b*u*v   this is the eqn for the predator population
```

u is the prey population number  
v is the predator population number  
a is the constant growth rate of the prey (w/o predation)  
b is the constant prey death rate due to predation  
c is the constant predator death rate (w/o prey as the food source)  
d is a constant that corresponds to how many prey a predator must consume to reproduce  

The array X will serve to represent the time-resolved growth rates resulting from the two ODEs above.  

```python
from numpy import *
import pylab as p

a = 1.
b = 0.1
c = 1.5
d = 0.75

def dX_dt(X, t=0):
    """ Return the growth rate of fox and rabbit populations. """
    return array([ a*X[0] -   b*X[0]*X[1] ,
                  -c*X[1] + d*b*X[0]*X[1] ])
```

We need to define the equilibrium points where the growth rate for the system will be zero, as defined below. At these points, the system can be linearized by taking the Jacobian matrix (A) of it, defined as d2X_dt2:

```python
X_f0 = array([     0. ,  0.])
X_f1 = array([ c/(d*b), a/b])
all(dX_dt(X_f0) == zeros(2) ) and all(dX_dt(X_f1) == zeros(2)) # => True

def d2X_dt2(X, t=0):
    """ Return the Jacobian matrix evaluated in X. """
    return array([[a -b*X[1],   -b*X[0]     ],
                  [b*d*X[1] ,   -c +b*d*X[0]] ])
```

X_f0 defines the point at which both species appear to be going to extinction (i.e. population growth goes to zero).  As predator number drops sufficiently, the prey population can rebound.
X_f1, when evaluated by the Jacobian, gives imaginary eigenvalues, representative of the periodic nature of this model.

```python
A_f0 = d2X_dt2(X_f0)                                                             
A_f1 = d2X_dt2(X_f1)                    
                                        
lambda1, lambda2 = linalg.eigvals(A_f1) 

T_f1 = 2*pi/abs(lambda1)                #this represents fluctuation period
```

Now we are ready to integrate our system of equations through scipy:

```python
from scipy import integrate
t = linspace(0, 15,  1000)              # dimensionless time
X0 = array([10, 5])                     # initials population numbers: 10 prey, 5 predator
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']                     # >>> 'Integration successful.'
```

Finally, we generate the plot that is incorporated in our original article:

```python
prey, predator = X.T
f1 = p.figure()
p.plot(t, prey, 'r-', label='Prey')
p.plot(t, predator  , 'b-', label='Predator')
p.grid()
p.legend(loc='best')
p.xlabel('time')
p.ylabel('population')
p.title('Dynamics of Predator-Prey Populations')
f1.savefig('ex.predprey1.png')
```

There you have it.  As is intuitive, increases in prey population lead to increases in predator population. 
Increased predation has a negative feedback effect on system population, and the cycle restarts.  

If we calculate T_f1 from Line 8, we indeed see that the period matches to about 5.1 time units as seen in the plot.
